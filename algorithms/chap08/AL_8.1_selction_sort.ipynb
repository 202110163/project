#### 문제정의 01)

거스름돈을 돌려주어야 할 때 최소 몇 개의 동전이 필요한지 구하는 문제

#### 알고리즘 설명02)

주어진 금액 V를 500,100,50,10,5,1 동전중 가장 적은 동전을 만드는 방법을 greedy 알고리즘을 사용

#### 손으로 푼 예제 03)

#### 코드개요04)

### 함수:min_coins_greedy(coins,V)
### 입력:동전의 종류 리스트  'coins' 와 목표 금액  'V'
### 출력:각 동전을 종류별로 필요한 개수 리스트 'count'

#### 코드05)

```python
def min_coins_geerdy(coins, V):
    count = []
    for i in range(len(coins)):
        cnt = V // coins[i]
        count.append(cnt)
        V -= cnt* coins[i]
    return count
```

#### 테스트 코드06)

```python
def min_coins_geerdy(coins, V):
    count = []
    for i in range(len(coins)):
        cnt = V // coins[i]
        count.append(cnt)
        V -= cnt* coins[i]
    return count
        
coins = [500,100,50,10,5,1]
V = 580
print("동전 액수 =",V)
print("동전 종류 =",coins)
print("동전 개수 =",min_coins_geerdy(coins,V))
```
